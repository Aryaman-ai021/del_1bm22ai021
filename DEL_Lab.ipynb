```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# XOR input and output
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Define the model
model = keras.Sequential()
model.add(layers.Dense(2, input_dim=2, activation='relu'))  # Hidden layer with 2 neurons
model.add(layers.Dense(1, activation='sigmoid'))            # Output layer

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=1000, verbose=0)

# Evaluate the model
predictions = model.predict(X)
predicted_classes = (predictions > 0.5).astype(int)

# Print the results
for i in range(len(X)):
    input_val = X[i]
    output_val = y[i][0]
    pred = predicted_classes[i][0]
    print(f"Input: {input_val} | Expected: {output_val} | Predicted: {pred}")
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# Example dataset (e.g., 3 classes)
X = np.array([[1, 2], [1, 4], [1, 0],
              [2, 2], [2, 4], [2, 0],
              [3, 2], [3, 4], [3, 0]])
y = np.array([[0], [0], [0],
              [1], [1], [1],
              [2], [2], [2]])

# One-hot encoding for the labels
y = keras.utils.to_categorical(y, num_classes=3)

# Define the model
model = keras.Sequential()
model.add(layers.Dense(4, input_dim=2, activation='relu'))  # Hidden layer
model.add(layers.Dense(3, activation='softmax'))            # Output layer

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=100, verbose=1)

# Make predictions
predictions = model.predict(X)
predicted_classes = np.argmax(predictions, axis=1)

# Print the results
for i, pred in enumerate(predicted_classes):
    print(f"Input: {X[i]} | Predicted Class: {pred}")
```
